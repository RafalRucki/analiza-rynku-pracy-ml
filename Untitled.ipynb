```python
import requests
import pandas as pd

# Ustawiamy adres do ich wewnętrznego silnika wyszukiwania (przykładowo dla Python)
# To jest często bardziej stabilne niż wyciąganie danych z wyglądu strony
url = "https://nofluffjobs.com/api/search/posting"

# Przykładowe parametry (szukamy ofert związanych z Pythonem)
payload = {
    "criteria": "python",
    "page": 1
}

response = requests.post(url, json=payload)

if response.status_code == 200:
    data = response.json()
    postings = data.get('postings', [])
    print(f"Sukces! Znaleziono {len(postings)} ofert pracy.")
    
    # Tworzymy prostą tabelkę z pierwszymi 5 ofertami
    jobs_list = []
    for job in postings[:10]:
        jobs_list.append({
            'Tytuł': job.get('title'),
            'Firma': job.get('name'),
            'Technologia': job.get('category'),
            'Lokalizacja': job.get('location', {}).get('places', [{}])[0].get('city', 'Remote')
        })
    
    df = pd.DataFrame(jobs_list)
    print(df)
else:
    print(f"Błąd połączenia: {response.status_code}. Może wymagają innych nagłówków?")
```
